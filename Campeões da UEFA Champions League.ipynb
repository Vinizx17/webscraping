```python
# Inicialize uma lista para armazenar os dados da tabela
table_data = []
```

```python
# Encontre todas as linhas (<tr>) na tabela
rows = table.find_all("tr")
```

```python
# Itere sobre as linhas da tabela
for row in rows:
    # Inicialize uma lista para armazenar os dados de uma linha
    row_data = []
    
    # Encontre todas as células (<th> ou <td>) na linha
    cells = row.find_all(["th", "td"])
    
    # Itere sobre as células da linha
    for cell in cells:
        # Extraia o texto da célula e adicione à lista de dados da linha
        row_data.append(cell.get_text(strip=True))
    
    # Adicione os dados da linha à lista de dados da tabela
    table_data.append(row_data)
```

```python
# Crie um DataFrame do Pandas com os dados da tabela
df = pd.DataFrame(table_data[1:], columns=table_data[0][:3])

display(df)
```
